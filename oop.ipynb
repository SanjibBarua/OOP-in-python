```python
class phone:
    category="Electronics"    ## attribute
## creating object
apple=phone()
xiaomi=phone()
realme=phone()
print(apple.category)
print(xiaomi.category)
print(realme.category)
```

### Instance attribute

```python
class phone:
    category="Electronics"    ## attribute
    ## constructor
    def __init__(self,model,battery,camera,battery_percentage=100):
        self.model=model
        self.battery=battery
        self.camera=camera
        self.battery_percentage=battery_percentage

    ## methods
    def charge(self,hour):
        self.charge +=hour
    def capture(self,total_photo):
        if(self.battery_percentage)<=0:
            print("no charge")
        else: 
            self.battery_percentage-=total_photo
            print("photo captured")
               
## creating object
apple=phone("iphone17",5000,40)
xiaomi=phone("redmi14",5000,50)
realme=phone("realmepro",3000,40)

print(apple.battery)
print(realme.battery)
print(apple.battery_percentage)
```
